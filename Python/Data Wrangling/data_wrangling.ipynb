# Data Normalization :
- uniform the data
- making sure they have same impact
- aik machli samundar mein aur aik jar mein
- also for computational reasons

```python
kashti.head()
```

```python
ks4 = kashti[['age in days', 'fare']]
ks4.head()
```

- The above data is really in wide range and we need to normalize because it is hard to compare
- Normalization changes the values to the range of 0-to-1 ( now both variable has similar influence on our models )
# Methods of Normalization
1 Simple feature scaling

    - x(new)= x(old)/x(max)

2 Min-Max method

3 Z-score ( standard score ) -3 to +3

4 Log transformation

# Simple feature scaling : 

```python
ks4['fare'] = ks4['fare']/ks4['fare'].max()
ks4['age in days'] = ks4['age in days']/ks4['age in days'].max()
ks4.head()
```

# Min-Max Method :

```python
ks4['fare'] = (ks4['fare']-ks4['fare'].min()) / (ks4['fare'].max()-ks4['fare'].min())
ks4.head()
```

# Z-score : ( standard core )

```python
ks4['fare'] = (ks4['fare']-ks4['fare'].mean()) / ks4['fare'].std()
ks4.head()
```
